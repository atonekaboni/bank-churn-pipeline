# Bank Customer Churn Analysis (EDA)
**Dataset:** 10,000 European bank customers account info.  

**Customer Churn:** number of customers lost over time.  
**Tenure:** the length of time a customer has maintained an active relationship with a business.  

## Import & QA the data

```python
import pandas as pd

churn_cust_info = pd.read_excel("dataset/Bank_Churn_Messy.xlsx")
churn_cust_info.head()
```

```python
churn_acct_info = pd.read_excel("dataset/Bank_Churn_Messy.xlsx", sheet_name = 1)
churn_acct_info.head()
```

```python
churn_df = churn_cust_info.merge(churn_acct_info, how="left", on="CustomerId")
churn_df.head()
```

```python
# Duplicate rows and columns
churn_df.info()
```

```python
# churn_df = churn_df.drop("Tenure_y", axis=1).rename({"Tenure_x":"Tenure"}, axis=1).drop_duplicates()
churn_df = churn_df.drop("Tenure_y", axis=1).rename({"Tenure_x":"Tenure"}, axis=1)
churn_df.head()
```

## Data Prep & Feature Engineering

```python
# Remove unnecessary columns for the model
churn_df.head()
```

```python
modelling_df = churn_df.drop(["CustomerId", "Surname"], axis = 1)
modelling_df.head()
```

```python
# drop first will reduce redundant column, ex. France column dropped here
pd.get_dummies(modelling_df, drop_first=True)
```

```python
modelling_df = pd.get_dummies(modelling_df, drop_first=True)
```

```python
modelling_df["Balance_v_Sal"] = modelling_df["Balance"] / modelling_df["EstimatedSalary"]
modelling_df.head()
```

```python
sns.boxplot(data=modelling_df, y="Balance_v_Sal")
```

```python
modelling_df.describe()
```

```python
# Remove outliers with a high balance-to-salary ratio
sns.boxplot(data=modelling_df.query("Balance_v_Sal < 10"), y="Balance_v_Sal")
```
